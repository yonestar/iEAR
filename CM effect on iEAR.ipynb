## Does CM lead to more present-moment focused emotions?  (seems like a long shot)

```python
tenses1 = ['Tense_C1', 'Tense_P1', 'Tense_F1', 'Tense_U1']
tenses2 = ['Tense_C2', 'Tense_P2', 'Tense_F2', 'Tense_U2']

for i, e in enumerate(tenses1):
    [r,p] = pearsonr(iear[e], iear[tenses2[i]])
    print 'test-retest for ' + e[0:-1] + ':  r = ' + str(round(r,2)) + ':  p = ' + str(round(p,2))
```

```python
plt.boxplot(iear[tenses1].as_matrix())
plt.xticks([1,2,3,4],['current', 'past', 'future', 'unclear'])
plt.figure()
plt.boxplot(iear[tenses2].as_matrix())
plt.xticks([1,2,3,4],['current', 'past', 'future', 'unclear'])
```

```python
iear['Tense_C%1'] = iear['Tense_C1'] / iear[tenses1].sum(axis=1)
iear['Tense_C%2'] = iear['Tense_C2'] / iear[tenses2].sum(axis=1)
fig=plt.boxplot([iear['Tense_C%1'], iear['Tense_C%2']])
plt.xticks([1,2],['time1', 'time2'])
plt.title('% of emotion that is current focused')
```

##### OK, looks like Ps mostly "felt" about the current moment, and some about the past.  little about the future.

### let's look at effects of interventions

```python
iear['Tense_C%diff'] = iear['Tense_C%2'] - iear['Tense_C%1']
sns.barplot('Group', 'Tense_C%diff',palette=palt,data=iear)
```

##### no effect observable

# Does CM increase other-focused emotions?

```python
objs1 = ['Obj_S1', 'Obj_W1', 'Obj_O1', 'Obj_N1', 'Obj_U2']
objs2 = ['Obj_S2', 'Obj_W2', 'Obj_O2', 'Obj_N2', 'Obj_U2']

for i, e in enumerate(objs1):
    [r,p] = pearsonr(iear[e], iear[objs2[i]])
    print 'test-retest for ' + e[0:-1] + ':  r = ' + str(round(r,2)) + ':  p = ' + str(round(p,2))
```

```python
plt.boxplot(iear[objs1].as_matrix())
plt.xticks([1,2,3,4,5],['self', 'we/us', 'other person', 'non-person', 'unclear'])
locs, labels = plt.xticks()
a=plt.setp(labels, rotation=30)

plt.figure()
plt.boxplot(iear[objs2].as_matrix())
plt.xticks([1,2,3,4,5],['self', 'we/us', 'other person', 'non-person', 'unclear'])
locs, labels = plt.xticks()
a=plt.setp(labels, rotation=30)
```

```python
iear['Obj_OW%1'] = (iear['Obj_O1'] + iear['Obj_W1']) / iear[objs1].sum(axis=1)
iear['Obj_OW%2'] = (iear['Obj_O2'] + iear['Obj_W2']) / iear[objs2].sum(axis=1)
fig=plt.boxplot([iear['Obj_OW%1'], iear['Obj_OW%2']])
plt.xticks([1,2],['time1', 'time2'])
plt.title('% of emotion that is focused on other ppl')
```

```python
iear['Obj_OW%diff'] = iear['Obj_OW%2'] - iear['Obj_OW%1']
sns.barplot('Group', 'Obj_OW%diff',palette=palt,data=iear)
```

```python
ftest('Obj_OW%diff')
```

##### effect in hypothesized direction, but not significant.

# Does CM affect how much time Ps spend with others?

```python
# remember that don't sum to 1 if nothing was transcribed but coder could still tell that P was alone or with others etc.
others1 = ['withoneother1', 'withothers1', 'alone1']
others2 = ['withoneother2', 'withothers2', 'alone2']

sns.boxplot(iear[others1])
plt.title('Time 1')
plt.figure()
sns.boxplot(iear[others2])
plt.title('Time 2')
```

```python
for i, e in enumerate(others1):
    [r,p] = pearsonr(iear[e], iear[others2[i]])
    print 'test-retest for ' + e[0:-1] + ':  r = ' + str(round(r,2)) + ':  p = ' + str(round(p,2))
```

OK, so underlying data has some high variability, but its reliable.  let's see how % with others looks

```python
iear['%withother/s1'] = iear[['withoneother1','withothers1']].sum(axis=1) / iear[others1].sum(axis=1)
iear['%withother/s2'] = iear[['withoneother2','withothers2']].sum(axis=1) / iear[others2].sum(axis=1)
sns.lmplot('%withother/s1','%withother/s2',data=iear,hue='Group')

plt.figure()
iear['%withother/s1'].hist()
plt.figure()
iear['%withother/s2'].hist()
```
